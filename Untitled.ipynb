```python
df.AVERAGE_ACCT_AGE.str.replace('yrs', '').str.replace('mon', '').str.split(' ', expand = True)
```

```python
df[['year', 'month']] = df.AVERAGE_ACCT_AGE.str.replace('yrs', '').str.replace('mon', '').str.split(' ', expand = True)
```

```python
df.head(1)
```

```python
df['AVERAGE_ACCT_AGE'] = df['month'].astype(int) + (df['year'].astype(int) * 12)
```

```python
df.head(5)
```

```python
df[['year_', 'month_']] = df.CREDIT_HISTORY_LENGTH.str.replace('yrs', '').str.replace('mon', '').str.split(' ', expand = True)
```

```python
df['CREDIT_HISTORY_LENGTH'] = df['month_'].astype(int) + (df['year_'].astype(int) * 12)
```
